```python
## Dataset 1: Diabetic Data Preprocessing

import pandas as pd
import numpy as np

df = pd.read_csv("../Data Collection/Datasets/diabetic_data.csv")

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
(df == '?').sum()
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
df.isnull().sum()
```

```python
(df == '?').sum()
```

```python
df.drop(columns=[ 'weight', 'payer_code','medical_specialty','max_glu_serum','A1Cresult'], inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='most_frequent')

df[['race', 'diag_1', 'diag_2', 'diag_3']] = imputer.fit_transform(
    df[['race', 'diag_1', 'diag_2', 'diag_3']]
)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.to_csv("Cleaned Datasets/cleaned_diabetic_data.csv", index=False)
```
